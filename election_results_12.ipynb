```python
#List unique values in the df_16['cand'] column
df_12clean.cand.unique()
```

```python
## create new column that is number of voters for obama
# Create a new column called df_12clean.democrat where the value is the votes if df_12clean.cand equals 'OBAMA'
# and 0 if not.
df_12clean['democrat'] = df_12clean['votes'].where(df_12clean['cand']=='OBAMA', 0)
df_12clean
```

```python
## create new column that is number of voters for romney
# Create a new column called df_12clean.republican where the value is the votes if df_12clean.cand equals 'ROMNEY' or 
# and 0 if not.
df_12clean['republican'] = df_12clean['votes'].where(df_12clean['cand']=='ROMNEY', 0)
df_12clean
```

```python
## create new column that is number of voters for any other candidate
# Create a new column called df_12clean.other where the value is the votes if df_12clean.cand equals 
#'STEIN', 'JOHNSON', 'WRITE-IN' and 0 if not.
df_12clean['other'] = df_12clean['votes'].where((df_12clean['cand']=='STEIN') | (df_12clean['cand']=='JOHNSON') | (df_12clean['cand']=='WRITE-IN'),0)
df_12clean
```

```python
#group by FIPS
grp = df_12clean.groupby(['countycode',"FIPS"])

#calculate totals by FIPS of democrat, republican, and other
df_12partyvotes = grp[['democrat','republican','other']].sum()
df_12partyvotes
```

```python
list(df_12partyvotes)
```

```python
# create new column with WGTP30/WGTP to get % renter hhds in poverty
df_12partyvotes['victory_margin'] = ((df_12partyvotes['republican'] - df_12partyvotes['democrat']))/((df_12partyvotes['republican'] + df_12partyvotes['democrat']))
df_12partyvotes.head()
```

```python
# export election results with VICTORY MARGINS by county in PA, 2012
df_12partyvotes.to_csv('electionresults12_PAcounties.csv')
```
