```python
## create new column that is number of voters for obama
# Create a new column called df_08clean.democrat where the value is the votes if df_08clean.cand equals 'OBAMA'
# and 0 if not.
df_08clean['democrat'] = df_08clean['votes'].where(df_08clean['cand']=='OBAMA', 0)
df_08clean
```

```python
## create new column that is number of voters for mccain
# Create a new column called df_08clean.republican where the value is the votes if df_08clean.cand equals 'MCCAIN' or 
# and 0 if not.
df_08clean['republican'] = df_08clean['votes'].where(df_08clean['cand']=='MCCAIN', 0)
df_08clean
```

```python
## create new column that is number of voters for any other candidate
# Create a new column called df_08clean.other where the value is the votes if df_08clean.cand equals 
#'NADER', 'BARR', 'WRITE-IN' and 0 if not.
df_08clean['other'] = df_08clean['votes'].where((df_08clean['cand']=='NADER') | (df_08clean['cand']=='BARR') | (df_08clean['cand']=='WRITE-IN'),0)
df_08clean
```

```python
#group by FIPS
grp = df_08clean.groupby(['countycode',"FIPS"])

#calculate totals by FIPS of democrat, republican, and other
df_08partyvotes = grp[['democrat','republican','other']].sum()
df_08partyvotes
```

```python
list(df_08partyvotes)
```

```python
# create new column with WGTP30/WGTP to get % renter hhds in poverty
df_08partyvotes['victory_margin'] = ((df_08partyvotes['republican'] - df_08partyvotes['democrat']))/((df_08partyvotes['republican'] + df_08partyvotes['democrat']))
df_08partyvotes.head()
```

```python
# export election results with VICTORY MARGINS by county in PA, 2008
df_08partyvotes.to_csv('electionresults08_PAcounties.csv')
```
